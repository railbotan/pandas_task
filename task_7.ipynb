```python
import pandas as pd
```

```python
data = pd.read_csv("./works.csv")
```

```python
# Распределение по зарплате с высшем образованием
data.query("gender == 'Мужской' and educationType == 'Высшее'").hist(bins=100, alpha=0.5)
data.query("gender == 'Женский' and educationType == 'Высшее'").hist(bins=100, alpha=0.5)
```

```python
# Распределение по зарплате с незаконченным высшем образованием
data.query("gender == 'Мужской' and educationType == 'Незаконченное высшее'").hist(bins=100, alpha=0.5)
data.query("gender == 'Женский' and educationType == 'Незаконченное высшее'").hist(bins=100, alpha=0.5)
```

```python
# Распределение по зарплате со средним образованием
data.query("gender == 'Мужской' and educationType == 'Среднее'").hist(bins=100, alpha=0.5)
data.query("gender == 'Женский' and educationType == 'Среднее'").hist(bins=100, alpha=0.5)
```

```python
# Распределение по зарплате со средним профессиональным образованием
data.query("gender == 'Мужской' and educationType == 'Среднее профессиональное'").hist(bins=100, alpha=0.5)
data.query("gender == 'Женский' and educationType == 'Среднее профессиональное'").hist(bins=100, alpha=0.5)
```
