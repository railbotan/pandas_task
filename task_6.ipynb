```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./works.csv")
percentiles = np.linspace(.1, 1, 10)

salary_men = data.query('gender == "Мужской"').quantile(percentiles)
salary_women = data.query('gender == "Женский"').quantile(percentiles)
```

```python
fig1, ax = plt.subplots()
ax.plot(percentiles, salary_men)
plt.xlabel("Перцентили")
plt.ylabel("Зарплата мужчин")

plt.show()

fig2, ax = plt.subplots()
ax.plot(percentiles, salary_men)
plt.xlabel("Перцентили")
plt.ylabel("Зарплата женщин")

plt.show()
```
