```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
works = pd.read_csv("./works.csv")
percentiles = np.linspace(.1, 1, 10)

salary_men = works.query('gender == "Мужской"').quantile(percentiles)
salary_women = works.query('gender == "Женский"').quantile(percentiles)
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(percentiles, salary_men)
plt.ylim([0, 800000])
plt.title("Зарплата мужчин по перцентилям")
plt.xlabel("Перцентили")
plt.ylabel("Зарплата мужчин")

plt.show()

fig2, ax2 = plt.subplots()
ax2.plot(percentiles, salary_women)
plt.ylim([0, 800000])
plt.title("Зарплаты женщин по перцентилям")
plt.xlabel("Перцентили")
plt.ylabel("Зарплата женщин")

plt.show()
```
